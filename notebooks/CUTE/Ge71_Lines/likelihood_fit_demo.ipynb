```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 100000
background = np.random.uniform(low=0.0, high=1.0, size=int(n/2))
signal = np.random.normal(loc=0.5, scale=0.1, size=int(n))
spectrum = signal #np.concatenate([background, signal])
```

```python
plt.figure(figsize=(4,3), dpi=200)
plt.hist(spectrum, bins = np.linspace(0, 1, 51))
plt.xlabel('Energy')
plt.ylabel('Count')
```

```python
bins = 50
NLL = np.zeros((bins, bins))
mean_guess = np.linspace(0.4, 0.6, bins)
std_guess = np.linspace(0.07, 0.25, bins)

for i, std in enumerate(std_guess):
    for j, mean in enumerate(mean_guess):
        NLL[i][j] = -(-n/2*np.log(2*np.pi) -n/2*np.log(std**2) -1/(2*std**2)*np.sum([(x-mean)**2 for x in spectrum]))
```

```python
plt.figure(figsize=(4,3), dpi=200)

X, Y = np.meshgrid(mean_guess, std_guess)
Z = NLL

plt.contour(X, Y, Z, levels = 50)
plt.axhline(0.1, 0, 1, ls = '--', color = 'black')
plt.axvline(0.5, 0, 1, ls = '--', color = 'black')

plt.xlabel(r'$\mathrm{\mu}$')
plt.ylabel(r'$\mathrm{\sigma}$')
```
