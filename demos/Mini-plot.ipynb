```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np                 
import pickle
from collections import defaultdict
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
def plot_averaged_traces_with_inset(averaged_traces, time_bins, title):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot the big plot first
    for chan_name, avg_trace in averaged_traces.items():
        ax.plot(time_bins, avg_trace, label=chan_name)
    
    ax.set_title(title)
    ax.set_xlim(-10, 1000)
    ax.set_xlabel(r"Time [$\mathrm{\mu s}$]")
    ax.set_ylabel(r"Amplitude [$\mathrm{\mu A}$]")
    ax.legend()

    # Create an inset of the zoomed-in plot
    ax_inset = inset_axes(ax, width="30%", height="30%", loc='upper right', borderpad=8)
    for chan_name, avg_trace in averaged_traces.items():
        ax_inset.plot(time_bins, avg_trace, label=chan_name)
    
    ax_inset.set_xlim(-10, 50)
    ax_inset.set_ylim(-0.20, max(max(avg_trace) for avg_trace in averaged_traces.values()) * 0.5)  # Adjust y limits
    ax_inset.set_title('Zoomed In')
    ax_inset.set_xlabel(r"Time [$\mathrm{\mu s}$]")
    ax_inset.set_ylabel(r"Amplitude [$\mathrm{\mu A}$]")

    plt.show()
```
